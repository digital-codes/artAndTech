## Basic Programming Elements
We start here to introduce some fundamental elements of programming
 * Functions
 * Loops
 * Conditions
 
 Functions are used to define code sections which are used repeatedly. Instead of writing the same code
 over and over again, we write the code in a specific paragraph with a name and *call* it via this name
 whenever we need the functionality. We can also provide parameters on every call, in this example the
 number of items to print and the color.
 
 Functions provide structure to our code and enable modular programming, as we can move functions into files
 which can improve code readability.
 
 
 Loops are another way to repeat stuff but at a different level: they actually control the execution of statements in the program. 
 There are different kinds of loops:
   * The *FOR* loop in the segment function repeats an action for a certain number of times
   * The *WHILE* loop in the main body reapeats an action while a certain *condition* is true
 
 Conditions are logic statements which resolve to either *True* or *False*, like *5 < 3* (which is False). The result is typically used to decide upon subsequent actions via *IF - ELSE* statemets or in *WHILE* loops.

And of course we have all the typcall mathematical and logical operators like 
> +, + , * , / , AND, OR, NOT

In this example we create  a simply checker-board type pattern by printing 0's or 1's.
Modify the constants and check the results.

```python
""" 
Demonstrate simple loops, conditional statements
and functions by creating a checker board
Size is N by N fields
Field size is configurable, same width and height
"""

########################################################################
# this is our first function declaration
# We have to declare the function before we use it
# This function is used to make one linear segment of a pattern
# we start with the definition which provides a name and 
# the parameters 
# The next line, with the triple quotes is the so called docstring
# and we use it to document the funtions operation and 
# parameters. The content of a docstring is returned with
# the "help" function, see below
def segment(width, color):
    """ print <width> items of a <color> """
    # add items with a FOR LOOP
    for i in range(width):
        #print(color,end="") # simple verion
        print(f"{color:2d}",end="") # this looks better

########################################################################
# After the function declaration we continue by ddecalring some
# constant and the main body of the program

# but first, let's call the help on our custom function
help(segment)

# constants (actually just variables we do not change any more)
FIELDS = 4
FIELD_WIDTH = 4

# variables
row = 0
column = 0
# set color to 0
color = 0
# we iterate until we are done ...
while row < FIELDS*FIELD_WIDTH and column < FIELDS*FIELD_WIDTH:
    # print segment
    segment(FIELD_WIDTH, color)
    # invert color after every segment
    color = 1 if color == 0 else 0
    # increment column. go back to 0 after 8 fields
    column = column + FIELD_WIDTH if column < FIELD_WIDTH*(FIELDS-1) else 0
    # new line and increment row if column = 0
    if column == 0:
        print("") # prints a new line
        # next row
        row = row + 1
    # invert color again if new row and end of field
    if column == 0 and row % FIELD_WIDTH == 0 :
        color = 1 if color == 0 else 0


########################################################################
```
